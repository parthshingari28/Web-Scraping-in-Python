# Web Scraping : extract data using url library & BeautifulSoup

```python
# Import packages
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
# Specify url of the web page
source = urlopen('https://en.wikipedia.org/wiki/COVID-19').read()
```

```python
# Make a soup 
soup = BeautifulSoup(source,'lxml')
soup
```

```python
print(set([text.parent.name for text in soup.find_all(text=True)]))
```

```python
# Extract the plain text content from paragraphs
text = ''
for paragraph in soup.find_all('p'):
    text += paragraph.text
text
```

```python
# Clean text
text = re.sub(r'\[.*?\]+', '', text)
text = text.replace('\n', '')
text
```
